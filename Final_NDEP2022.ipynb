```python
def dictionary_builder(x_train, x_test, y_train, y_test):
  """creates a dictionary containing ten keys (0 through 9), where the value of each
  key ia a 3D array containing all the 7000 or so images associated with that key."""
  import numpy as np
  images = np.concatenate((x_train, x_test), axis=0)
  labels = np.concatenate((y_train, y_test))
  image_dictionary = {}
  for label in range(10):
    image_dictionary[label] = images[labels == label]
  return images, labels, image_dictionary

```

The following cell contains a function that evaluates the effect of the number of nodes in the hiddenlayer of a NN with a single hidden layer on the classifier performance. The classifier is trained to recognize hanwritten digits.  

```python
def nn_evaluator_1(image_dictionary):
  """evaluates the effect of the number of nodes in the hidden layer of a NN
   with a single hidden layer on the classifier performance.
   The classifier is trained to recognize hanwritten digits."""
  n_trainers = 5000 #number of trainers from each class
  epochs = 5
  # n_hiddenLayerNodes = [15, 30, 60, 120, 240]
  n_hiddenLayerNodes = [5, 10, 15, 20, 25, 30, 40, 50, 100]
  import numpy as np
  import tensorflow as tf
  #
  def nn_evaluator_1_sub1(x_train, x_test):
    """Normalize the train and test datasets."""
    x_train = tf.keras.utils.normalize(x_train, axis=1)
    x_test = tf.keras.utils.normalize(x_test, axis=1)
    return x_train, x_test
  #
  def nn_evaluator_1_sub2(n_nodes):
    """Build the model object consisting of input-layer with 784 nodes (flatten-layer) ,output-layer
     with 10 nodes, and a single hidden-layer with n_nodes. After building the model it compiles it. """
    model = tf.keras.models.Sequential()
    # Add the Flatten Layer
    model.add(tf.keras.layers.Flatten())
    # Build the the hidden layers
    model.add(tf.keras.layers.Dense(n_nodes, activation=tf.nn.relu))
    # Build the output layer
    model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))
    # Compile the model
    model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model
  #
  def nn_evaluator_1_sub3(x_train, y_train, model, epochs):
    """Trains the model."""
    model.fit(x=x_train, y=y_train, epochs=5)
    return model
  #
  def nn_evaluator_1_sub4(model, x_test, y_test):
    """Evaluate the performance of the trained model using untrained-on test images."""
    test_loss, test_acc = model.evaluate(x=x_test, y=y_test)
    return test_loss, test_acc
  #
  def nn_evaluator_1_sub5(image_dictionary, n_trainers):
    """It picks the trainers from each class randomly and uses the remaining images for testing.
    n_trainers is the number of trainers from each class."""
    for digit in range(10):
      digit_images = image_dictionary[digit]
      n_images = digit_images.shape[0]
      v1 = np.random.permutation(n_images)
      digit_trainers = digit_images[v1[:n_trainers]]
      digit_testers = digit_images[v1[n_trainers:]]
      digit_trainer_labels = digit * np.ones(n_trainers).astype(np.uint8)
      digit_tester_labels = digit * np.ones(n_images-n_trainers).astype(np.uint8)
      if digit == 0:
        x_train = digit_trainers
        x_test = digit_testers
        y_train = digit_trainer_labels
        y_test = digit_tester_labels
      else:
        x_train = np.concatenate((x_train, digit_trainers))
        x_test = np.concatenate((x_test, digit_testers))
        y_train = np.concatenate((y_train, digit_trainer_labels))
        y_test = np.concatenate((y_test, digit_tester_labels))
    return x_train, x_test, y_train, y_test
    #
  x_train, x_test, y_train, y_test = nn_evaluator_1_sub5(image_dictionary, n_trainers)
  x_train, x_test = nn_evaluator_1_sub1(x_train, x_test)
  #
  accuracyRecord = []
  for j_hiddenLayerNodes in n_hiddenLayerNodes:
    model = nn_evaluator_1_sub2(j_hiddenLayerNodes)
    model = nn_evaluator_1_sub3(x_train, y_train, model, epochs)
    test_loss, test_acc = nn_evaluator_1_sub4(model, x_test, y_test)
    accuracyRecord.append(test_acc)
  return accuracyRecord






```

```python
def nn_evaluator_2(image_dictionary):
  """evaluates the effect of the number of trainers on the performance of the NN
  with a single hidden layer with fixed number of nodes. For each setting of the number of trainers,
  the training and testing is done multiple times (n_runs). This is important if the number of trainers
  is low (i.e. less than a thousand) so that the NN performance is not affected by the peculiarity
  of the training set which may be overly affected by few bad examples.
  The classifier is trained to recognize hanwritten digits."""
  n_hiddenLayerNodes = 100
  epochs = 5
  # n_trainers = [50, 100, 200, 300, 400, 500, 1000, 2000, 3000, 4000, 5000]
  # n_uns = 7 #
  n_trainers = [5, 100, 200, 300, 400, 500]
  n_runs = 5
  import numpy as np
  import tensorflow as tf
  #
  def nn_evaluator_2_sub1(x_train, x_test):
    """Normalize the train and test datasets."""
    x_train = tf.keras.utils.normalize(x_train, axis=1)
    x_test = tf.keras.utils.normalize(x_test, axis=1)
    return x_train, x_test
  #
  def nn_evaluator_2_sub2(n_nodes):
    """Build the model object consisting of input-layer with 784 nodes (flatten-layer) ,output-layer
     with 10 nodes, and a single hidden-layer with n_nodes. After building the model it compiles it. """
    model = tf.keras.models.Sequential()
    # Add the Flatten Layer
    model.add(tf.keras.layers.Flatten())
    # Build the the hidden layers
    model.add(tf.keras.layers.Dense(n_nodes, activation=tf.nn.relu))
    # Build the output layer
    model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))
    # Compile the model
    model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model
  #
  def nn_evaluator_2_sub3(x_train, y_train, model, epochs):
    """Trains the model."""
    model.fit(x=x_train, y=y_train, epochs=5)
    return model
  #
  def nn_evaluator_2_sub4(model, x_test, y_test):
    """Evaluate the performance of the trained model using untrained-on test images."""
    test_loss, test_acc = model.evaluate(x=x_test, y=y_test)
    return test_loss, test_acc
  #
  def nn_evaluator_2_sub5(image_dictionary, n_trainers):
    """It picks the trainers from each class randomly and uses the remaining images for testing.
    n_trainers is the number of trainers from each class."""
    for digit in range(10):
      digit_images = image_dictionary[digit]
      n_images = digit_images.shape[0]
      v1 = np.random.permutation(n_images)
      digit_trainers = digit_images[v1[:n_trainers]]
      digit_testers = digit_images[v1[n_trainers:]]
      digit_trainer_labels = digit * np.ones(n_trainers).astype(np.uint8)
      digit_tester_labels = digit * np.ones(n_images-n_trainers).astype(np.uint8)
      if digit == 0:
        x_train = digit_trainers
        x_test = digit_testers
        y_train = digit_trainer_labels
        y_test = digit_tester_labels
      else:
        x_train = np.concatenate((x_train, digit_trainers))
        x_test = np.concatenate((x_test, digit_testers))
        y_train = np.concatenate((y_train, digit_trainer_labels))
        y_test = np.concatenate((y_test, digit_tester_labels))
    return x_train, x_test, y_train, y_test
    #
  def nn_evaluator_2_sub6(image_dictionary, n_trainers, n_hiddenLayerNodes, epochs):
    """computes the classification accuracy for one instantiation of the experiment."""
    x_train, x_test, y_train, y_test = nn_evaluator_2_sub5(image_dictionary, n_trainers)
    x_train, x_test = nn_evaluator_2_sub1(x_train, x_test)
    model = nn_evaluator_2_sub2(n_hiddenLayerNodes)
    model = nn_evaluator_2_sub3(x_train, y_train, model, epochs)
    test_loss, test_acc = nn_evaluator_2_sub4(model, x_test, y_test)
    return test_acc
    #
  # n0 = n_trainers[0]
  # accuracyRecordMatrix = [[]]
  # accuracyRecordList = []
  for j_trainers in n_trainers:
    print('j_trainers = ', j_trainers)
    accuracyRecord = []
    for j_run in np.arange(n_runs):
      test_acc = nn_evaluator_2_sub6(image_dictionary, j_trainers, n_hiddenLayerNodes, epochs)
      accuracyRecord.append(test_acc)
    accuracyRecord = np.array(accuracyRecord)
    accuracyRecord = np.reshape(accuracyRecord, (1, len(accuracyRecord)))
    print('accuracyRecord = ', accuracyRecord)
    # accuracyRecord = np.reshape(accuracyRecord, (len(accuracyRecord), 1)
    if j_trainers == n_trainers[0]:
      accuracyRecordMatrix = accuracyRecord
    else:
      accuracyRecordMatrix = np.concatenate((accuracyRecordMatrix,accuracyRecord), axis=0)
  #   accuracyRecordList.append(accuracyRecord)
  # accuracyRecordMatrix = np.array(accuracyRecordList)
  # return accuracyRecordList
  return accuracyRecordMatrix


```

```python
def cnn_evaluator_5(image_dictionary, n_trainers):
  """Evaluates and compares the performance of three different convolutional neural networks (CNN).
  Network-one has one conv-pooling layer and one fully-connected (FC) layer.
  Network-two has two conv-pooling layers and two fully-connected (FC) layers.
  Network-three has three conv-pooling layers and three fully-connected (FC) layers.
  All three networks are followed by the output layer which has ten nodes."""
  #
  cnnNetworkType = [1, 2, 3]
  epochs = 5
  #
  import numpy as np
  import tensorflow as tf
  #
  def cnn_evaluator_5_sub1(x_train, x_test):
    """Normalize the train and test datasets."""
    x_train = tf.keras.utils.normalize(x_train, axis=1)
    x_test = tf.keras.utils.normalize(x_test, axis=1)
    return x_train, x_test
  #
  def cnn_evaluator_5_sub2(network_type):
    """Depending of the input parameter network_type, which can be 1, 2, or 3, this function
    forms a CNN with one, two, or three conv/pooling layers followed by one, two, or three FC layers."""
    from keras.models import Sequential
    from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
    from keras.optimizers import gradient_descent_v2
    from keras.metrics import categorical_crossentropy

    model = Sequential()
    if network_type == 1: #one conv/pooling layer and one FC layer
      model.add(Conv2D(filters = 6, kernel_size = (5,5), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Flatten())
      model.add(Dense(120, activation = 'relu'))
      #
    elif network_type == 2:#two conv/pooling layers and two FC layers
      model.add(Conv2D(filters = 6, kernel_size = (5,5), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Conv2D(filters = 12, kernel_size = (5,5), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Flatten())
      model.add(Dense(120, activation = 'relu'))
      model.add(Dense(60, activation = 'relu'))
      #
    else: #three conv/pooling layers and three FC layer
      model.add(Conv2D(filters = 6, kernel_size = (5,5), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Conv2D(filters = 12, kernel_size = (5,5), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Conv2D(filters = 24, kernel_size = (5,5), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Flatten())
      model.add(Dense(120, activation = 'relu'))
      model.add(Dense(60, activation = 'relu'))
      model.add(Dense(30, activation = 'relu'))
      #
    # Add the output layer
    model.add(Dense(10, activation = 'softmax'))
    model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model
  #
  def cnn_evaluator_5_sub3(x_train, y_train, model, epochs):
    """Trains the model."""
    model.fit(x=x_train, y=y_train, epochs=5)
    return model
  #
  def cnn_evaluator_5_sub4(model, x_test, y_test):
    """Evaluate the performance of the trained model using untrained-on test images."""
    test_loss, test_acc = model.evaluate(x=x_test, y=y_test)
    return test_loss, test_acc
  #
  def cnn_evaluator_5_sub5(image_dictionary, n_trainers):
    """It picks the trainers from each clothing-category randomly and uses the remaining images for testing.
    n_trainers is the number of trainers from each category."""
    for category in range(10):
      category_images = image_dictionary[category]
      n_images = category_images.shape[0]
      v1 = np.random.permutation(n_images)
      category_trainers = category_images[v1[:n_trainers]]
      category_testers = category_images[v1[n_trainers:]]
      category_trainer_labels = category * np.ones(n_trainers).astype(np.uint8)
      category_tester_labels = category * np.ones(n_images-n_trainers).astype(np.uint8)
      if category == 0:
        x_train = category_trainers
        x_test = category_testers
        y_train = category_trainer_labels
        y_test = category_tester_labels
      else:
        x_train = np.concatenate((x_train, category_trainers))
        x_test = np.concatenate((x_test, category_testers))
        y_train = np.concatenate((y_train, category_trainer_labels))
        y_test = np.concatenate((y_test, category_tester_labels))
    return x_train, x_test, y_train, y_test
    #
  def cnn_evaluator_5_sub6(image_dictionary, n_trainers, epochs, network_type):
    """computes the classification accuracy for one instantiation of the experiment."""
    x_train, x_test, y_train, y_test = cnn_evaluator_5_sub5(image_dictionary, n_trainers)
    x_train, x_test = cnn_evaluator_5_sub1(x_train, x_test)
    model = cnn_evaluator_5_sub2(network_type)
    model = cnn_evaluator_5_sub3(x_train, y_train, model, epochs)
    test_loss, test_acc = cnn_evaluator_5_sub4(model, x_test, y_test)
    print('test_acc=', test_acc)
    return test_acc
  #
  for network_type in cnnNetworkType:
    print('network_type=', network_type)
    accuracyList = []
    for j_trainers in n_trainers:
      print('j_trainers=', j_trainers)
      accuracy = cnn_evaluator_5_sub6(image_dictionary, j_trainers, epochs, network_type)
      accuracyList.append(accuracy)
    accuracyList = np.array(accuracyList)
    accuracyList = np.expand_dims(accuracyList, axis=0)
    if network_type == cnnNetworkType[0]:
      accuracyMatrix = accuracyList
    else:
      accuracyMatrix = np.concatenate((accuracyMatrix, accuracyList), axis=0)
  return accuracyMatrix



```

```python
def cnn_evaluator_6(image_dictionary, n_trainers):
  """Evaluates and compares the performance of three different convolutional neural networks (CNN).
  Network-one has one conv-pooling layer (6@5x5) and one fully-connected (FC) layer.
  Network-two has two conv-pooling layers(6@5x5), (12@9x9) and two fully-connected (FC) layers.
  Network-three has three conv-pooling (6@5x5), (12@8x8), (24@11x11) layers and three fully-connected (FC) layers.
  All three networks are followed by the output layer which has ten nodes."""
  #
  cnnNetworkType = [1, 2, 3]
  epochs = 5
  #
  import numpy as np
  import tensorflow as tf
  #
  def cnn_evaluator_6_sub1(x_train, x_test):
    """Normalize the train and test datasets."""
    x_train = tf.keras.utils.normalize(x_train, axis=1)
    x_test = tf.keras.utils.normalize(x_test, axis=1)
    return x_train, x_test
  #
  def cnn_evaluator_6_sub2(network_type):
    """Depending of the input parameter network_type, which can be 1, 2, or 3, this function
    forms a CNN with one, two, or three conv/pooling layers followed by one, two, or three FC layers."""
    from keras.models import Sequential
    from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
    from keras.optimizers import gradient_descent_v2
    from keras.metrics import categorical_crossentropy

    model = Sequential()
    if network_type == 1: #one conv/pooling layer and one FC layer
      model.add(Conv2D(filters = 6, kernel_size = (5,5), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Flatten())
      model.add(Dense(120, activation = 'relu'))
      #
    elif network_type == 2:#two conv/pooling layers and two FC layers
      model.add(Conv2D(filters = 6, kernel_size = (5,5), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Conv2D(filters = 12, kernel_size = (8,8), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Flatten())
      model.add(Dense(120, activation = 'relu'))
      model.add(Dense(60, activation = 'relu'))
      #
    else: #three conv/pooling layers and three FC layer
      model.add(Conv2D(filters = 6, kernel_size = (5,5), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Conv2D(filters = 12, kernel_size = (8,8), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Conv2D(filters = 24, kernel_size = (11,11), padding = 'same',
                   activation = 'relu', input_shape = (28,28,1)))
      model.add(MaxPooling2D(pool_size = (2,2)))
      model.add(Flatten())
      model.add(Dense(120, activation = 'relu'))
      model.add(Dense(60, activation = 'relu'))
      model.add(Dense(30, activation = 'relu'))
      #
    # Add the output layer
    model.add(Dense(10, activation = 'softmax'))
    model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model
  #
  def cnn_evaluator_6_sub3(x_train, y_train, model, epochs):
    """Trains the model."""
    model.fit(x=x_train, y=y_train, epochs=5)
    return model
  #
  def cnn_evaluator_6_sub4(model, x_test, y_test):
    """Evaluate the performance of the trained model using untrained-on test images."""
    test_loss, test_acc = model.evaluate(x=x_test, y=y_test)
    return test_loss, test_acc
  #
  def cnn_evaluator_6_sub5(image_dictionary, n_trainers):
    """It picks the trainers from each clothing-category randomly and uses the remaining images for testing.
    n_trainers is the number of trainers from each category."""
    for category in range(10):
      category_images = image_dictionary[category]
      n_images = category_images.shape[0]
      v1 = np.random.permutation(n_images)
      category_trainers = category_images[v1[:n_trainers]]
      category_testers = category_images[v1[n_trainers:]]
      category_trainer_labels = category * np.ones(n_trainers).astype(np.uint8)
      category_tester_labels = category * np.ones(n_images-n_trainers).astype(np.uint8)
      if category == 0:
        x_train = category_trainers
        x_test = category_testers
        y_train = category_trainer_labels
        y_test = category_tester_labels
      else:
        x_train = np.concatenate((x_train, category_trainers))
        x_test = np.concatenate((x_test, category_testers))
        y_train = np.concatenate((y_train, category_trainer_labels))
        y_test = np.concatenate((y_test, category_tester_labels))
    return x_train, x_test, y_train, y_test
    #
  def cnn_evaluator_6_sub6(image_dictionary, n_trainers, epochs, network_type):
    """computes the classification accuracy for one instantiation of the experiment."""
    x_train, x_test, y_train, y_test = cnn_evaluator_6_sub5(image_dictionary, n_trainers)
    x_train, x_test = cnn_evaluator_6_sub1(x_train, x_test)
    model = cnn_evaluator_6_sub2(network_type)
    model = cnn_evaluator_6_sub3(x_train, y_train, model, epochs)
    test_loss, test_acc = cnn_evaluator_6_sub4(model, x_test, y_test)
    print('test_acc=', test_acc)
    return test_acc
  #
  for network_type in cnnNetworkType:
    print('network_type=', network_type)
    accuracyList = []
    for j_trainers in n_trainers:
      print('j_trainers=', j_trainers)
      accuracy = cnn_evaluator_6_sub6(image_dictionary, j_trainers, epochs, network_type)
      accuracyList.append(accuracy)
    accuracyList = np.array(accuracyList)
    accuracyList = np.expand_dims(accuracyList, axis=0)
    if network_type == cnnNetworkType[0]:
      accuracyMatrix = accuracyList
    else:
      accuracyMatrix = np.concatenate((accuracyMatrix, accuracyList), axis=0)
  return accuracyMatrix



```

```python
def cnn_evaluator_7(image_dictionary, n_trainers):
  """Trains a CNN with one conv/pooling layer and one FC layer. (incomplete)
  It can be trained with either the MNIST-digit dataset or the MNIST-fashion dataset.
  The output is the trained CNN model as well as the test-images and the corresponding labels.
  The trained model can subsequently be used to assign labels to any input image including the test images.
  The outputs of this function are later used to compute the confusion matrix associated with the
  trained model. n_trainers are the number of training images from each class."""
  #
  epochs = 5
  #
  import numpy as np
  import tensorflow as tf
  #
  def cnn_evaluator_7_sub1(x_train, x_test):
    """Normalize the train and test datasets."""
    x_train = tf.keras.utils.normalize(x_train, axis=1)
    x_test = tf.keras.utils.normalize(x_test, axis=1)
    return x_train, x_test
  #
  def cnn_evaluator_7_sub2():
    from keras.models import Sequential
    from keras.layers import Conv2D, MaxPooling2D, Dense, Flatten
    from keras.optimizers import gradient_descent_v2
    from keras.metrics import categorical_crossentropy

    model = Sequential()
    model.add(Conv2D(filters = 6, kernel_size = (5,5), padding = 'same',
                  activation = 'relu', input_shape = (28,28,1)))
    model.add(MaxPooling2D(pool_size = (2,2)))
    model.add(Flatten())
    model.add(Dense(120, activation = 'relu'))
    model.add(Dense(10, activation = 'softmax'))
    model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
    return model
  #
  def cnn_evaluator_7_sub3(x_train, y_train, model, epochs):
    """Trains the model."""
    model.fit(x=x_train, y=y_train, epochs=5)
    return model
  #
  def cnn_evaluator_7_sub4(model, x_test):
    """Evaluate the assigned classes to each os the test images."""
    p_testImages = x_test.shape[0]
    assignedClassList = []
    for j_test_image in range(p_testImages):
      testImage = x_test[j_test_image]
      testImage = testImage.flatten()
      assignedClass = model.predict_classes(testImage)
      assignedClassList.assignedClassList
    return assignedClassList
  #
  def cnn_evaluator_7_sub5(image_dictionary, n_trainers):
    """It picks the trainers from each clothing-category randomly and uses the remaining images for testing.
    n_trainers is the number of trainers from each category."""
    for category in range(10):
      category_images = image_dictionary[category]
      n_images = category_images.shape[0]
      v1 = np.random.permutation(n_images)
      category_trainers = category_images[v1[:n_trainers]]
      category_testers = category_images[v1[n_trainers:]]
      category_trainer_labels = category * np.ones(n_trainers).astype(np.uint8)
      category_tester_labels = category * np.ones(n_images-n_trainers).astype(np.uint8)
      if category == 0:
        x_train = category_trainers
        x_test = category_testers
        y_train = category_trainer_labels
        y_test = category_tester_labels
      else:
        x_train = np.concatenate((x_train, category_trainers))
        x_test = np.concatenate((x_test, category_testers))
        y_train = np.concatenate((y_train, category_trainer_labels))
        y_test = np.concatenate((y_test, category_tester_labels))
    return x_train, x_test, y_train, y_test
    #
  x_train, x_test, y_train, y_test = cnn_evaluator_7_sub5(image_dictionary, n_trainers)
  x_train, x_test = cnn_evaluator_7_sub1(x_train, x_test) #normalizes the training and test images
  model = cnn_evaluator_7_sub2() #builds the model archtecture
  model = cnn_evaluator_7_sub3(x_train, y_train, model, epochs) #computes the trained model
  return x_train, x_test, y_train, y_test, model
```

```python
def confusionMatrixComputer(image_dictionary, n_trainers):
  """Computes the confusion matimage_dirix associated with the CNN classifier
  obtained from cnn_evaluator_7(image_dictionary, n_trainers).
  image_dictionary can be either the dictionary associated with the MNIST digits dataset of the
  MNIST fashion dataset.
  The output confusionMatrix is a 10x10-matrix, where each row contains the probabilitirs associated with
  a typical image from the corresponding class being labeled as member of any of the ten classes by the trained classifier.
  The trained classifier is computed by cnn_evaluator_7 function in the above cell, which is called by the current function."""
  #
  #computes the trained model (classifier) and lists all the test images, which have not been used to train the classifier,
  #and their corresponding labels (classes)
  x_train, x_test, y_train, y_test, model = cnn_evaluator_7(image_dictionary, n_trainers)
  #
  #computes the outputs (responses) of the trained classifier to all the test images.
  #The output to each test image is a vector of ten numbers. Each row of ClassifierOutMatrix
  #represents the response of the classifier to one test image.
  ClassifierOutMatrix = model.predict(x_test)
  #
  #determines the classes assigend to each test image by the classifier
  imageClassesComputed = np.argmax(ClassifierOutMatrix, axis=1)
  #
  #determine the indices of ll the misclassified test images
  misclassified_indices = np.where(imageClassesComputed != y_test)[0]

  # return ClassifierOutMatrix, imageClassesComputed, misclassified_indices
  # exit()
  #
  for class_true in range(10):
    z1 = np.where(y_test == class_true)[0]
    n1 = len(z1)
    z2 = imageClassesComputed[z1]
    mat_row = np.zeros((1, 10))
    for class_computed in range(10):
      z3 = np.sum(z2 == class_computed)
      z4 = 100 * z3/n1
      mat_row[0, class_computed] = z4
    if class_true == 0:
      confusion_matrix = mat_row
    else:
      confusion_matrix = np.concatenate((confusion_matrix, mat_row), axis=0)

  return ClassifierOutMatrix, imageClassesComputed, misclassified_indices, model, x_test, y_test, confusion_matrix


```
