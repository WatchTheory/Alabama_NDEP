```python
def dictionary_builder(x_train, x_test, y_train, y_test):
  """creates a dictionary containing ten keys (0 through 9), where the value of each
  key ia a 3D array containing all the 7000 or so images associated with that key."""
  import numpy as np
  images = np.concatenate((x_train, x_test), axis=0)
  labels = np.concatenate((y_train, y_test))
  image_dictionary = {}
  for label in range(10):
    image_dictionary[label] = images[labels == label]
  return images, labels, image_dictionary

```

```python
images, labels, image_dictionary = dictionary_builder(x_train, x_test, y_train, y_test)
```

Build a tiled image where each column contains images from the same class.

```python
def mosaic_maker(image_dictionary, n_rows):
  for j_column in range(10):
    x1 = np.random.permutation(5000)[0:n_rows]
    for j_row in range(n_rows):
      x2 = image_dictionary[j_column][x1[j_row]]
      if j_row == 0:
        column_tile = x2
      else:
        column_tile = np.concatenate((column_tile, x2), axis=0)
    if j_column == 0:
      mosaic_image = column_tile
    else:
      mosaic_image = np.concatenate((mosaic_image, column_tile), axis=1)
  fig = plt.gcf()
  fig.set_size_inches(18.5, 15.5)
  ax = plt.gca()
  ax.axes.xaxis.set_visible(False)
  ax.axes.yaxis.set_visible(False)
  plt.imshow(mosaic_image, cmap='gray')
  return mosaic_image

# n_rows = 20; mosaic_image = mosaic_maker(image_dictionary, n_rows)
```

```python
import tensorflow as tf
```

Preprocess the data

```python
# Normalize the train dataset
x_train = tf.keras.utils.normalize(x_train, axis=1)
# Normalize the test dataset
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

Build the model

```python
#Build the model object
model = tf.keras.models.Sequential()
# Add the Flatten Layer
model.add(tf.keras.layers.Flatten())
# Build the input and the hidden layers
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(128, activation=tf.nn.relu))
# Build the output layer
model.add(tf.keras.layers.Dense(10, activation=tf.nn.softmax))
```

Compile the model

```python
# Compile the model
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

Train the model

```python
model.fit(x=x_train, y=y_train, epochs=5) # Start training process
```

Evaluate the model

```python
# Evaluate the model performance
test_loss, test_acc = model.evaluate(x=x_test, y=y_test)
# Print out the model accuracy
print('\nTest accuracy:', test_acc)
```

Make predictions

```python
predictions = model.predict([x_test]) # Make prediction
```
